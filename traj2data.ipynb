```python
from ase.io import read, write
import os
from glob import glob
```

```python

from torch_geometric.data import Data
import torch
from tqdm import tqdm
def atoms2data(atoms):
    data= Data()
    data.pos = torch.as_tensor(atoms.get_positions(), dtype=torch.float)
    data.z = torch.as_tensor(atoms.get_atomic_numbers(), dtype=torch.long)
    data.energy = torch.as_tensor(atoms.get_potential_energy(), dtype=torch.float).view(-1)
    data.force = torch.as_tensor(atoms.get_forces(), dtype=torch.float)
    data.natoms = torch.tensor(len(atoms), dtype=torch.long)
    data.cell = torch.as_tensor(atoms.get_cell(), dtype=torch.float)
    return data
subdir_list=os.listdir('Path/to/your/trajectories')  # Replace with your actual path
for subdir in subdir_list:
    data_list=[]
    file_list=glob(f'Path/to/your/trajectories/{subdir}/*.traj')
    print(f"Processing {subdir} with {len(file_list)} files.")
    for file in tqdm(file_list,total=len(file_list)):
        atoms_list=read(file,index='1:')
        data_list.extend([atoms2data(atoms) for atoms in atoms_list])
    torch.save(data_list, f'Path/to/your/pt_file/{subdir}.pt')
    print(f"Saved {len(data_list)} data points to Path/to/your/pt_file/{subdir}.pt")
print("All files processed and saved.")
        
```
